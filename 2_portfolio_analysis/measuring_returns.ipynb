# Measuring Returns

## 1. Arithmetic Returns

```python
def calc_arithmetic_return(ev, sv) -> float:
    """
    Calculate the arithmetic return for the given parameters.

    Keyword arguments:
    ev -- End Value (default 0.0)
    sv -- Start Value (default 0.0)

    return:
    av -- Calculated arithmetic value
    """
    av = (ev-sv)/(sv)
    return av
```

```python
factor = calc_arithmetic_return(13_000_000, 10_000_000)
print(f"Portfolio increased by factor: {factor}")
```

Imagine we have following Scenario and you need to calculate the corresponding arithmetic returns for it:
> Start Value: 8_000_000 <br>
> Year 0 - Year 1: 15 % <br>
> Year 1 - Year 2: -15 % <br>

```python
sv = 8_000_000
ev_y1 = sv * 1.15 # +15 %
ev_y2 = ev_y1 * 0.85 # -15 %
```

```python
ev_y2
```

#### ❌ Arithmetic Returns Are Not Time-Additive

Arithmetic returns can mislead when evaluating multi-period investments:

- They **assume each year's return applies to the same base**, ignoring compounding.
- Adding annual returns like `+15%` and `−15%` may suggest a net return of `0%`, but this is incorrect.
- In volatile markets, this distortion becomes more pronounced.

## 1. Logarithmic Returns

```python
import math
math.log(ev_y1/sv) + math.log(ev_y2/ev_y1)
sv * (1-0.022756987122616312)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ANNUAL_CLOSING_VALUES = [
    35_000_000, # Year 0
    40_000_000, # Year 1
    37_000_000, # Year 2
    32_000_000, # Year 3
    34_000_000, # Year 4
    42_000_000, # Year 5
]
ANNUAL_CLOSING_VALUES_NP = np.array(ANNUAL_CLOSING_VALUES)
```

```python
ANNUAL_CLOSING_VALUES_NP[:-1]
```

```python
arithmetic_returns = (ANNUAL_CLOSING_VALUES_NP[1:] - ANNUAL_CLOSING_VALUES_NP[:-1]) / ANNUAL_CLOSING_VALUES_NP[:-1]
plt.plot(figsize=(10, 6))
plt.plot([2001,2002,2003,2004,2005],arithmetic_returns, label='Arithmetic Returns', marker='o')
```

```python
log_returns = np.log(ANNUAL_CLOSING_VALUES_NP[1:] / ANNUAL_CLOSING_VALUES_NP[:-1])
```

```python
plt.plot(log_returns, label='Logarithmic Returns', marker='x')
plt.title('Annual Arithmetic vs. Logarithmic Returns')
plt.xlabel('Year')
plt.ylabel('Returns')
plt.legend()
plt.grid(True)
plt.show()
```

```python
cumulative_arithmetic_return = np.cumsum(arithmetic_returns)
cumulative_logarithmic_return = np.exp(np.cumsum(log_returns)) - 1

plt.figure(figsize=(10, 6))
plt.plot(cumulative_arithmetic_return, label='Cumulative Arithmetic Returns', marker='o')
plt.plot(cumulative_logarithmic_return, label='Cumulative Logarithmic Returns', marker='x')
plt.title('Cumulative Arithmetic vs. Logarithmic Returns')
plt.xlabel('Year')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid(True)
plt.show()
```

```python
apple['Log_Ret'] = np.log(apple['Close'] / apple['Close'].shift(1))
ann_ret = apple['Log_Ret'].mean() * 252
print(ann_ret)
```
